```python
#Moving Average Plot

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('Dataset.csv')

# Convert Date to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Calculate moving averages
df['MA10'] = df['Close'].rolling(window=10).mean()
df['MA20'] = df['Close'].rolling(window=20).mean()

# Plot moving averages
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Close'], label='Close Price', color='blue')
plt.plot(df['Date'], df['MA10'], label='10-Day MA', color='red')
plt.plot(df['Date'], df['MA20'], label='20-Day MA', color='green')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Moving Averages')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Box plot for Price Distribution

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('Dataset.csv')

# Box plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['Open', 'High', 'Low', 'Close']])
plt.title('Price Distribution')
plt.show()
```

```python
#Scatterplot for percent change vs Volume

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('Dataset.csv')

# Inspect the first few rows to understand the data
print(df.head())

# Remove any leading or trailing spaces from column names
df.columns = df.columns.str.strip()

# Convert Percent Change to numeric (remove % and convert to float)
# First replace any non-numeric values with NaN, then convert to float
df['Percent Change'] = df['Percent Change'].str.replace('%', '').str.strip()
df['Percent Change'] = pd.to_numeric(df['Percent Change'], errors='coerce')

# Convert Volume to numeric (remove commas and any non-numeric characters)
df['Volume'] = df['Volume'].str.replace(',', '').str.strip()
df['Volume'] = pd.to_numeric(df['Volume'], errors='coerce')

# Drop rows with NaN values in 'Percent Change' or 'Volume' columns
df.dropna(subset=['Percent Change', 'Volume'], inplace=True)

# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['Percent Change'], df['Volume'], color='purple')
plt.xlabel('Percent Change')
plt.ylabel('Volume')
plt.title('Percent Change vs Trading Volume')
plt.grid(True)
plt.show()
```
